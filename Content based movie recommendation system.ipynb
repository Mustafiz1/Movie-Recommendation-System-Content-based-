## Column 'genres' modification

```python
movie.iloc[0].genres
```

```python
def convert(obj):
    L = []
    for i in obj:
        L.append(i['name'])
    return L
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
# making a list of dictionaries
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movie['genres'] = movie['genres'].apply(convert)
```

```python
movie.head()
```

```python
movie.keywords = movie.keywords.apply(convert)
```

```python
movie.head()
```

```python
p = ast.literal_eval(movie.cast[0])
p[0]['name']
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movie.cast = movie.cast.apply(convert3)
```

```python
movie.head()
```

```python
movie.crew[0]
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movie.crew  = movie.crew.apply(fetch_director)
```

```python
movie.head()
```

```python
movie.overview[0]
```

```python
movie.overview = movie.overview.apply(lambda x:x.split())
```

## Remove spaces between names (for same first or last name)

```python
# 'Sam Worthington'
# 'SamWorthington'
```

```python
movie.genres = movie.genres.apply(lambda x:[i.replace(' ', '') for i in x])
movie.keywords = movie.keywords.apply(lambda x:[i.replace(' ', '') for i in x])
movie.cast = movie.cast.apply(lambda x:[i.replace(' ', '') for i in x])
movie.crew = movie.crew.apply(lambda x:[i.replace(' ', '') for i in x])
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    return ' '.join(y)
```

## Distance Calculation (Cosine)

```python
# Better for multidimensional cases
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
similarity[1]
```

```python
new_df[new_df['title'] == 'Spectre'].index[0]
```

```python
sorted(similarity[0], reverse = True)
```

```python
list(enumerate(similarity[0]))
```

```python
sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x: x[1])[1:6]
```

```python
def recommend(m):
    m_index = new_df[new_df['title'] == m].index[0]
    distances = similarity[m_index]
    m_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x:x[1])[1:6]
    
    for i in m_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
